```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.types import *
engine = create_engine('mysql+pymysql://root:123456@127.0.0.1:3306/info?charset=utf8')
file_data=pd.read_sql('宁波二手房',engine)
file_data
```

```python
file_data.duplicated()
```

```python
file_data=file_data.drop_duplicates()
file_data
```

```python
file_data=file_data.dropna()
file_data
```

```python
import numpy as np
data_new=np.array([])
data=file_data['面积'].values
for i in data:
    data_new=np.append(data_new,np.array(i[:-2]))
data=data_new.astype(np.float64)
file_data.loc[:,'面积']=data
file_data
```

```python
housetype_data=file_data['户型']
temp_list=[]
for i in housetype_data:
    new_info=i.replace('房间','室')
    temp_list.append(new_info)
file_data.loc[:,'户型']=temp_list
file_data
```

```python
temp_list=[]
type_data=file_data['标签']
for i in type_data:
    new_info=i.replace(',','')
    temp_list.append(new_info)
file_data.loc[:,'标签']=temp_list
data_new=np.array([])
data=file_data['标签'].values
for i in data:
    data_new=np.append(data_new,np.array(i[:-3]))
data=data_new.astype(np.float64)
file_data.loc[:,'标签']=data
file_data
```

```python
import pandas as pd
new_df=pd.DataFrame({'区域':file_data['区域'].unique(),
                     '数量':[0]*11})
new_df
```

```python
file_data['位置']='宁波市'+file_data['区域'].values+file_data['小区'].values
file_data
```

```python
import requests
cun=[]
AK = "pYPV7iDVu4GGhoqCC0qdCnIuEfydlRrx"
for i in range(len(file_data['位置'])):
    name=file_data['位置'][i]
    url = 'http://api.map.baidu.com/geocoding/v3/?address=%s&output=json&ak=%s'%(name,AK)
    res = requests.get(url)
    if res.status_code==200:
        val = res.json()
        if val["status"]==0:
            retval = {'lng':val['result']['location']['lng'],'lat':val['result']['location']['lat'],'count':60}
            cun.append(retval)
            print(retval)
```

```python
with open('经纬度数据.txt', 'w') as f:
    for i in range(len(cun)):
        f.write(","+str(cun[i]))
print('写入完成')
```

```python
def all_house(arr):
    arr=np.array(arr)
    key=np.unique(arr)
    result={}
    for k in key:
        mask=(arr==k)
        arr_new=arr[mask]
        v=arr_new.size
        result[k]=v
    return result
house_array=file_data['户型']
house_info=all_house(house_array)
house_info
```

```python
house_type=dict((key,value) for key,value
               in house_info.items() if value > 50)
show_houses=pd.DataFrame({'户型':[x for x in house_type.keys()],
                        '数量':[x for x in house_type.values()]})
show_houses
```

```python
import matplotlib.pyplot as plt
import matplotlib
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False 
house_type=show_houses['户型']
house_type_num=show_houses['数量']
plt.barh(range(9),house_type_num,height=0.7,
         color='steelblue',alpha=0.8)
plt.yticks(range(9),house_type)
plt.xlim(0,2500)
plt.xlabel("数量")
plt.ylabel("户型种类")
plt.title("宁波地区各户型房屋数量")
for x,y in enumerate(house_type_num):
    plt.text(y+0.2,x-0.1,'%s' %y)
plt.show()
```

```python
df_all=pd.DataFrame({'区域':file_data['区域'].unique(),
                    '房价总金额':[0]*11,
                    "总面积":[0]*11})
df_all
```

```python
sum_price=file_data['价格信息（万/套）'].groupby(file_data['区域']).sum()
sum_area=file_data['面积'].groupby(file_data['区域']).sum()
df_all['房价总金额']=sum_price.values
df_all['总面积']=sum_area.values
df_all
```

```python
df_all['每平米方价格（万）']=round(df_all['房价总金额']/df_all['总面积'],2)
df_all
```

```python
df_merge=pd.merge(new_df,df_all)
df_merge
```

```python
import matplotlib.ticker as mtick
from matplotlib.font_manager import FontProperties
num=df_merge['数量']
price=df_merge['每平米方价格（万）']
l=[i for i in range(11)]
plt.rcParams["font.sans-serif"] = ["SimHei"]
lx=df_merge['区域']
fig=plt.figure()
ax1=fig.add_subplot(111)
ax1.plot(l,price,'or-',label='价格')
for i,(_x,_y) in enumerate(zip(l,price)):
    plt.text(_x,_y,price[i],color='black',fontsize=10)
ax1.set_ylim([0,10])
ax1.set_ylabel('价格')
plt.legend(prop={'family':'SimHei','size':8},loc='upper left')
ax2=ax1.twinx()
plt.bar(l,num,alpha=0.3,color='green',label='数量')
ax2.set_ylabel('数量')
ax2.set_ylim([0,1500])
plt.legend(prop={'family':'SimHei','size':8},loc="upper right")
plt.xticks(l,lx)
plt.show()
```
