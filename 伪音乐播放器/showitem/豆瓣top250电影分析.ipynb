```python
from lxml import etree
import requests
import csv 
data=open('IMDb.csv','wt',newline='',encoding='utf-8')
writer=csv.writer(data)
# 写入表头信息
writer.writerow(('rank','title','description','runtime','genre','rating','metascore','votes','gross_earning_in_mil','director','actor'))

#  构造urls
urls=['https://www.imdb.com/search/title/?title_type=feature&release_date=2017-01-01,2017-12-31&start={}'.format(i) for i in range(0,250,50)]

#  加入请求头
headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1;WOW64) AppleWebKit/537.36 (KHTML,like GeCKO) Chrome/45.0.2454.85 Safari/537.36 115Broswer/6.0.3'
}

for url in urls:
    # 用requests库获取网页信息,lxml解析html文件
    html=requests.get(url,headers=headers)
    selector=etree.HTML(html.text)
    infos=selector.xpath('//div[@class="lister-item mode-advanced"]')
    for info in infos:
        rank=info.xpath('div[3]/h3/span[1]/text()')
        title=info.xpath('div[3]/h3/a/text()')
        description=[i.replace("\n","").strip() for i in info.xpath('div[3]/p[2]/text()')]
        runtime=[i.replace("\n","").strip() for i in info.xpath('div[3]/p[1]/span[@class="runtime"]/text()')]
        genre=[i.replace("\n","").strip() for i in info.xpath('div[3]/p[1]/span[@class="genre"]/text()')]
        rating=[i.replace("\n","").strip() for i in info.xpath('div[3]/div/div/strong/text()')]
        metascore=info.xpath('normalize-space(div[3]/div/div[3]/span/text())')
        votes=info.xpath('div[3]/p[4]/span[2]/text()')
        gross_earning_in_mil=info.xpath('div[3]/p[4]/span[5]/text()')
        director=info.xpath('div[3]/p[3]/a[1]/text()')
        actors=[]
        stress=info.xpath('div[3]/p[3]')
        for i in stress:
            temp=i.xpath('a[position()>1]/text()')
            actors.append(temp)    
        print(rank,title,description,runtime,genre,rating,metascore,votes,gross_earning_in_mil,director,actors)
        writer.writerow((rank,title,description,runtime,genre,rating,metascore,votes,gross_earning_in_mil,director,actors))
data.close()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%config InlineBackend.figure_format = 'retina'
%matplotlib auto
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False 
df=pd.read_csv("IMDb.csv")
# 将空值形式的缺失值转换成可识别的类型
df['metascore'] =df['metascore'].fillna('0')
df['metascore'] =df['metascore'].astype(float)
df.head(20)
```

```python
df['rank'] =df['rank'].str.replace('[', '')
df['rank'] =df['rank'].str.replace(']', '')
df['title'] =df['title'].str.replace('[', '')
df['title'] =df['title'].str.replace(']', '')
df['runtime'] =df['runtime'].str.replace('[', '')
df['runtime'] =df['runtime'].str.replace(']', '')
df['genre'] =df['genre'].str.replace('[', '')
df['genre'] =df['genre'].str.replace(']', '')
df['rating'] =df['rating'].str.replace('[', '')
df['rating'] =df['rating'].str.replace(']', '')
df['votes'] =df['votes'].str.replace('[', '')
df['votes'] =df['votes'].str.replace(']', '')
df['gross_earning_in_mil'] =df['gross_earning_in_mil'].str.replace('[', '')
df['gross_earning_in_mil'] =df['gross_earning_in_mil'].str.replace(']', '')
df['director'] =df['director'].str.replace('[', '')
df['director'] =df['director'].str.replace(']', '')
df['actor'] =df['actor'].str.replace('[', '')
df['actor'] =df['actor'].str.replace(']', '')
df.head(20)
```

```python
df['gross_earning_in_mil'] =df['gross_earning_in_mil'].fillna('$0.00M')
df.head(20)
```

```python
df['main_genre'] = df['genre'].str.split(', ').str.get(0)
df.head()
```

```python
df['runtime'] = df['runtime'].str.split(' ').str.get(0).astype(int)
df.head()
```

```python
df.groupby(['gross_earning_in_mil']).mean()
```

```python
df.groupby(['gross_earning_in_mil','rank']).mean()
```

```python
# 获得电影类型的分布和平均得分数据
genre_data = df.groupby('main_genre')['title', 'rating'].agg({'title': 'count', 'rating': 'mean'}).sort_values(by='title', ascending=False)
# 绘制不同电影类型的数量分布
fig, ax2 = plt.subplots(figsize=(14,7))
plt.title('不同类型的IMDB电影数量', fontsize=18)

genre_data['title'].plot(kind='bar', width=0.8, color=(255/255,187/255,120/255))

# 设置坐标轴格式
ax2.set_xlabel('')
ax2.set_xticklabels(genre_data.index, rotation=0, fontsize=12)
ax2.tick_params(axis='x', rotation=0)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

# 标记不同电影类型的数量数据
x2 = np.arange(df['main_genre'].nunique())
y2 = df['main_genre'].value_counts().values
bar=ax2.bar(x2,y2,color='#CCFFFF')
y3 = genre_data['rating'].values
line=ax2.plot(x2,y3,color='#FF6633')
#添加无指向型文本
def autolabel(rects):
    for rect in rects:
        height =rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2,height+1,
                s='{}'.format(height),
                ha='center',va='bottom')
autolabel(bar)
#添加图例
ax2.legend([bar,line[0]],['数量','评分'],shadow=True,fancybox=True)
```
