```python
import pandas as pd
data=pd.read_csv('./credit_card.csv',encoding='gbk')
print('原始数据的形状为:',data.shape)
data
```

```python
#去除逾期记录为1
index1=(data["逾期"]==1) & (data["呆账"]==1) & (data["强制停卡记录"]==1) & \
(data["退票"]==1)& (data["拒往记录"]==1)& (data["瑕疵户"]==2)
data1=data.loc[index1==False,:]
print('删除异常记录后数据的形状为\n',data1.shape)
```

```python
data1
```

```python
index2=(data["呆账"]==1) & (data["强制停卡记录"]==1) & (data["退票"]==1)& \
(data["拒往记录"]==2)
data2=data1.loc[index2==False,:]
print('删除异常记录后数据的形状为\n',data2.shape)
```

```python
data2
```

```python
index3=(data["频率"]==5) & (data["月刷卡额"]!=1) 
data3=data2.loc[index3==False,:]
print('删除异常记录后数据的形状为\n',data3.shape)
```

```python
data3
```

```python
data1_selection=data3[["瑕疵户","逾期","呆账","强制停卡记录","退票","拒往记录"]]
data1_selection
```

```python
data2_selection=data3[["借款余额","个人月收入","个人月开销","家庭月收入","月刷卡额"]]
data2_selection
```

```python
data3_selection=data3[["职业","年龄","住家"]]
data3_selection
```

```python
#特征合并
card_features = pd.concat([data1_selection,data2_selection,data3_selection],axis=1)
print('构建LFS特征前5个为：\n',card_features.head())
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
card_features_Scaler=StandardScaler().fit_transform(card_features)
print('标准化后LPFMC模型的3个特征为:\n',card_features_Scaler)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5,random_state=123,algorithm="elkan").fit(card_features_Scaler)
```

```python
print("根据三个特征聚类后的聚类中心：",kmeans.cluster_centers_)
```

```python
num = pd.Series(kmeans.labels_).value_counts()
print('每类用户数目为：\n',num)
```
